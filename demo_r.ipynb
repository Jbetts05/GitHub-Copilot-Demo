# GitHub Copilot

Install both the <b>'GitHub Copilot Extension'</b> and the <b>'Github Copilot Chat Extension'</b>

Further more you will need to install the <b> R <b> extension for VScode, and within R, you will need to install the <b> IRkernel </b> package.

Following this, you would need to execute the IRkernel::installspec() function within R, to register the kernel with Jupyter.

There are three ways to use Copilot within VScode:
<br>

- ' # ' using comments, we instruct Copilot to generate code and docs
- ' Ctrl + i ' will open the Copilot Chat function within the cell, here you can enter your instruction to Copilot  
- You can also use Copilot Chat directly from your extension list, this can also be toggled with ' Ctrl + Alt + i '

```r
# using datasets and dplyr packages, let's put the iris dataset in a dataframe called df

# install.packages("datasets")
# install.packages("dplyr")
library(datasets)
library(dplyr)
df <- datasets::iris




```

# let's do a scatter plot of the features vs the species

```r
# install.packages("ggplot2")
library(ggplot2)

ggplot(df, aes(x = Sepal.Length, y = Sepal.Width, color = Species)) +
    geom_point() +
    labs(title = "Scatter plot of Sepal Length vs Sepal Width by Species")





```
